## Exercise 1  
1. Создайте график $x^2⋅sin(1/x^2)+x$ на интервале $[−1,1]$, используя 250 точек. Не забудьте пометить оси!
2. Создайте semilogy график (с логарифмической шкалой для оси y и линейной шкалой для оси x) относительной разницы  функций $1/(1+x^2)$ и $1/x^2$ на интервале $[5,25]$. (Относительная разница чисел $a$ и $b$ определяется как $|1−a/b|$)
3. В Scipy также есть много менее распространенных специальных функций. Взгляните на обширный список <a href=https://docs.scipy.org/doc/scipy/reference/special.html> здесь</a>! Выберите несколько интересных спец функций, которые вы знаете (или, может быть, не знаете), подставьте несколько чисел и начертите их.


```python
# Question 1

import numpy as np
import matplotlib.pyplot as plt

# Задаем интервал и количество точек
x = np.linspace(-1, 1, 250)

# Вычисляем значения функции
y = x**2 * np.sin(1 / x**2) + x

# Создаем график
plt.plot(x, y)

# Добавляем подписи к осям
plt.xlabel('x')
plt.ylabel('y')

# Показываем график
plt.show()
```

```python
# Question 2

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(5, 25, 100)  # Создание равномерно распределенных значений на интервале [5, 25]
y1 = 1 / (1 + x**2)  # Вычисление функции 1/(1+x^2)
y2 = 1 / x**2  # Вычисление функции 1/x^2

relative_diff = np.abs(1 - y1/y2)  # Вычисление относительной разницы

plt.semilogy(x, relative_diff)  # Построение semilogy графика
plt.xlabel('x')  # Подпись оси x
plt.ylabel('Относительная разница')  # Подпись оси y
plt.title('Относительная разница функций')  # Заголовок графика
plt.grid(True)  # Включение сетки
plt.show()  # Отображение графика
```

```python
# Exercise 3

import numpy as np
import matplotlib.pyplot as plt
import scipy

xx = np.linspace(-5, 5, 80)
yy = scipy.special.gamma(xx) / 25 # на максимум делить не хочет, поэтому нормировка условная
yy2 = scipy.special.psi(xx) / 50
yy3 = scipy.special.erf(xx)
plt.title('График 3')
plt.xlabel(u'Ox', family='verdana')
plt.ylabel(u'Oy', family='verdana')
plt.plot(xx, yy, label = 'нормированная гамма функция')
plt.plot(xx, yy2, label = 'нормированная пси функция', color = 'orange', linestyle = '--')
plt.plot(xx, yy3, label = 'функция ошибки', color = 'pink', linestyle = '', marker = '+')
plt.legend()
```

## Exercise 2  
1. Построить график функции $sin(x)/x$ на отрезке $[-6\pi;6\pi]$ и добавить тики по оси $x$ с шагом $\pi/2$ от $-6\pi$ до $6\pi$, причём обозначения тиков должны содержать греческую букву $\pi$. Добавить сетку на график.

```python
import numpy as np
import matplotlib.pyplot as plt

# Создание данных для графика
x = np.linspace(-6 * np.pi, 6 * np.pi, 1000)
y = np.sin(x) / x

# Создание графика
plt.plot(x, y, label='sin(x)/x')

# Добавление тиков на оси x с обозначениями, содержащими греческую букву π
ticks = np.linspace(-6 * np.pi, 6 * np.pi, 13)
plt.xticks(ticks, ['$-6\pi$', '$-5\pi$', '$-4\pi$', '$-3\pi$', '$-2\pi$', '$-\pi$', '$0$', '$\pi$', '$2\pi$', '$3\pi$', '$4\pi$', '$5\pi$', '$6\pi$'])

# Добавление сетки на график
plt.grid(True)

# Добавление легенды
plt.legend()
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
# Отображение графика
plt.show()

```

## Exercise 3  
1. Создайте цветовой график $sin(x)sin(y)$ на интервале $[−π,π]×[−π,π]$.
2. Создайте цветовой график $e^{-x^2-y^2}$ на интервале $[−3,3]×[−3,3]$.

```python
# Question 1

import numpy as np
import matplotlib.pyplot as plt

# Задаем интервал и количество точек на сетке
x = np.linspace(-np.pi, np.pi, 100)
y = np.linspace(-np.pi, np.pi, 100)

# Создаем сетку точек
X, Y = np.meshgrid(x, y)

# Вычисляем значения функции sin(x) * sin(y)
Z = np.sin(X) * np.sin(Y)

# Построение цветового графика
plt.figure(figsize=(8, 8))
plt.pcolormesh(X, Y, Z, cmap='RdYlBu')
plt.colorbar(label='Значение функции')
plt.title('График sin(x) * sin(y)')
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Создание сетки значений x и y
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)

# Вычисление значений функции
Z = np.exp(-X**2 - Y**2)

# Построение цветового графика
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, Z, cmap='viridis')
plt.colorbar(label='Value')
plt.title('Graph of exp(-x^2 - y^2)')
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
plt.show()
```
