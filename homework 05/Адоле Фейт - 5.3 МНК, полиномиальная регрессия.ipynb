## 1. Полиномиальная регрессия (5 баллов) 

Воспользуемся описанным свойством и будем находить приближение функции многочленом, решая систему линейных уравнений (1). Сформируйте систему линейных уравнений (то есть задайте матрицу коэффициентов A  и вектор правых частей b) для многочлена, который должен совпадать с функцией f в n точках, эквидистантно расположенных на отрезке x∈[1,15] (для этого удобно использовать функцию np.linspace()). Решите данную систему с помощью функции np.linalg.solve(). 

Постройте графики 1) исходной функции, 2) точек, по которым строился многочлен, 3) полиномиальную аппроксимацию. 

Хорошо ли полином приближает исходную функцию? 

Рассмотрите несколько различных значений n. Как меняется приближение с ростом числа точек (и соответственно степени полинома)? Хорошо ли ведёт себя полином при n=5 ? А при n=15? 

```python
import numpy as np
import numpy.linalg
import numpy.random
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-1/2 * x)


def approximation(n):
    a = np.linspace(1, 15, n)
    A = np.array([[x ** i for i in range(n)] for x in a])
    B = f(a)
    W = np.linalg.solve(A, B)
    answer = np.array([sum([W[i] * x**i for i in range(n)]) for x in a])
    return answer


xx = np.linspace(1, 15, 250)
n = 5
xx2 = np.linspace(1, 15, n)

# Graph 1: Original Function
fig, axs = plt.subplots(1, 3, figsize=(18, 4))

axs[0].grid()
axs[0].set_xlabel('x')
axs[0].set_ylabel('f(x)')
axs[0].plot(xx, f(xx), color='peru', label='Original Function')
axs[0].legend()

# Graph 2: Data Points
axs[1].grid()
axs[1].set_xlabel('x')
axs[1].set_ylabel('f(x)')
axs[1].plot(xx2, f(xx2), marker='.', color='green', linestyle='', label=f'Data Points (n={n})')
axs[1].legend()

# Graph 3: Approximation
axs[2].grid()
axs[2].set_xlabel('x')
axs[2].set_ylabel('f(x)')
axs[2].plot(xx2, approximation(n), color='red', label=f'Approximation (n={n})')
axs[2].plot(xx2, f(xx2), marker='.', color='green', linestyle='', label=f'Data Points (n={n})')
axs[2].legend()

plt.show()
```

## 3. Метод наименьших квадратов (5 баллов)

Третья часть задания посвящена собственно методу наименьших квадратов.
Выше мы строили полиномиальную регрессию той степени, которая соответствовала числу точек. Т.е. при количестве точек равных n мы получали полином степени n-1. На практике обычно ограничиваются полиномами степени не выше третьей, независимо от количества имеющихся точек.
Рассмотрим теперь случай, когда точек много. Пусть количество точек равно N, а полином по-прежнему имеет вид w_0+w_1 x+w_2 x^2+w_3 x^3+...+w_(n-1) x^(n-1), причём N>n. Тогда в  системе линейных уравнений вида (1) количество уравнений будет больше, чем число неизвестных. Подобного рода система уравнений может не иметь точного решения, особенно в случае, когда правая часть является приближённой, т.е. с шумом.
В таких случаях, вместо точного решения системы уравнений обычно решают задачу минимизации невязки уравнений:
(min)┬x ‖Ax-b‖ (2)
где в качестве нормы ‖⋅‖ обычно используют квадратичную норму. Поэтому такой подход называют методом наименьших квадратов.
Для решения данной задачи (2) используйте готовую функцию np.linalg.lstsq(), которая аналогично функции  np.linalg.solve() принимает два аргумента: матрицу коэффициентов A  и вектор правых частей b. Данная функция возвращает несколько объектов, первый из которых есть решение задачи.
Возьмите N=100 точек на отрезке x∈[1,15] и посчитайте в них значение функции f, добавьте гауссов с большим шумом с σ=0.1. Постройте методом наименьших квадратов кубический полином по данным точкам.
Постройте графики 1) исходной функции, 2) точек, по которым строился многочлен, 3) полиномиальную аппроксимацию.
Сделайте выводы по всей работе.


```python
import numpy as np
import numpy.linalg
import numpy.random
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-1/2 * x)

def add_noise(y, sigma=0.1):
    noise = np.random.normal(0, sigma, len(y))
    return y + noise

def approximation(n):
    a = np.linspace(1, 15, n)
    A = np.array([[x ** i for i in range(n)] for x in a])
    B = add_noise(f(a))
    W = np.linalg.lstsq(A, B, rcond=None)[0]
    xx = np.linspace(1, 15, 250)
    answer = np.array([sum([W[i] * x**i for i in range(n)]) for x in xx])
    return answer


N = 100
n = 4  # Cubic polynomial: n = 4, including constant term

xx = np.linspace(1, 15, 250)
xx_points = np.linspace(1, 15, N)

# Generate points with added noise
yy_points = add_noise(f(xx_points), sigma=0.1)

# Cubic polynomial approximation using least squares
approximation_curve = approximation(n)

# Graphs
fig, axs = plt.subplots(1, 3, figsize=(18, 4))

# Graph 1: Original Function
axs[0].grid()
axs[0].set_xlabel('x')
axs[0].set_ylabel('f(x)')
axs[0].plot(xx, f(xx), color='peru', label='Original Function')
axs[0].legend()

# Graph 2: Data Points with Noise
axs[1].grid()
axs[1].set_xlabel('x')
axs[1].set_ylabel('f(x)')
axs[1].plot(xx_points, yy_points, marker='.', color='green', linestyle='', label=f'Data Points with Noise (N={N})')
axs[1].legend()

# Graph 3: Approximation
axs[2].grid()
axs[2].set_xlabel('x')
axs[2].set_ylabel('f(x)')
axs[2].plot(xx, approximation_curve, color='red', label=f'Approximation (n={n})')
axs[2].plot(xx_points, yy_points, marker='.', color='green', linestyle='', label=f'Data Points with Noise (N={N})')
axs[2].legend()

plt.show()
```
