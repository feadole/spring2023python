```python
img = Image.open('peacock.jpg') #.realize
```

```python
img
```

```python
x = np.array(img)
```

```python
x.shape
```

```python
def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray
```

```python
Image.fromarray(np.asarray(rgb2gray(x), dtype=np.uint8))
```

```python
y = x.mean(axis=2)
y
```

```python
y.shape
```

```python
U, S, V = np.linalg.svd(y, full_matrices=False)
```

```python
V.shape
```

```python
U.shape
```

```python
S.shape
```

```python
Y_full = U.dot(np.diag(S)).dot(V)
Image.fromarray(np.asarray(Y_full,dtype=np.uint8))
```

 ## вывод

Разложения низкого ранга на основе SVD позволяют сжимать изображения. Усекая сингулярные значения и выбирая разные значения ранга сжатия, вы можете контролировать степень сжатия и качество реконструированных изображений. Понижение ранга приводит к потере деталей и ухудшению качества, в то время как повышение ранга сохраняет больше информации и улучшает качество. Оптимальный выбор значения ранга требует компромисса между качеством изображения и коэффициентом сжатия, как мы можем видеть в наших кодах выше.
